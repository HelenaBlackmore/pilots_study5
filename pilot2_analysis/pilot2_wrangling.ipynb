# DATA CLEANING

```python
# libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# loading data

pd.set_option('display.max_rows', 500)

pilot_raw=pd.read_csv("data\pilot2_raw.csv")
consent_summary = pd.read_csv("data\consent_summary.csv")
ddt_summary = pd.read_csv("data\ddt_summary.csv")
data_IP = pd.read_csv('data/batch_pilot2.iqx_participantlog.csv')
data_email = pd.read_csv('data/consent_survey.csv')
```

```python
# formatting data
pilot_raw = pilot_raw[['script.subjectid','trialnum', 'trialcode', 'response', 'correct', 'stimulusitem1', 'stimulusitem2' ]]
consent_summary = consent_summary[['script.subjectid']]
ddt_summary = ddt_summary[['script.subjectid', 'values.k']]

data_IP = data_IP[['SubjectId', 'ipaddress']]
data_IP['subject'] = data_IP['SubjectId']
data_IP = data_IP.drop('SubjectId', axis = 1)
data_IP = data_IP.dropna() #removes missing values (just tidying messy csv file)
data_IP.subject = data_IP.subject.astype("int64") # subject ids are inetgers

data_email = data_email[['subject', 'age_response', 'email_response', 'name_response']]
```

```python
data_IP['subject'].nunique() #no of pp who started the study
data_IP['ipaddress'].nunique() # unique IP addresses (started study)
IPs = data_IP['ipaddress'].values.tolist()
def Repeat(x):
    _size = len(x)
    repeated = []
    for i in range(_size):
        k = i + 1
        for j in range(k, _size):
            if x[i] == x[j] and x[i] not in repeated:
                repeated.append(x[i])
    return repeated

duplicate_IP = Repeat(IPs)

len(Repeat(IPs))

#print(("{} pp started the study, out of these, tehre were {} unique IP addresses and {} repeat IP addresses.").format(data_IP['subject'].nunique(), data_IP['ipaddress'].nunique(), len(Repeat(IPs))))
```

```python
#write a file with duplicate IPs:
textfile = open("duplicateIP.txt", "w")

for element in duplicate_IP:
    textfile.write(('{} \n').format(element))
textfile.close()
```

```python
# only keep unique IP addresses, removes all duplicates (keeps last entry)

unique_IP= data_IP.drop_duplicates(subset='ipaddress', keep='last')

ID_un_IP= unique_IP.subject.values.tolist() # list of pp first completion
len(ID_un_IP) #103




unique_IP
```

```python
data1 = pd.merge(
    pilot_raw,
    consent_summary,
    how="inner",
    on='script.subjectid',
    sort=True)

data_all = pd.merge(
    data1,
    ddt_summary,
    how="inner",
    on='script.subjectid',
    sort=True)
```

```python

data_all['script.subjectid'].nunique() #56 completed surveys

# using unique IP variable, get only unique responses:

data_all['subject']=data_all['script.subjectid']

# ?merge? 

data_un =pd.merge(data_all['subject'], unique_IP, on = 'subject')
data_un['subject'].nunique() # 44 unique responses (unique IP addresses)

```

```python
#now I need the email addresses as a list 

data_personal = pd.merge(data_email, data_un, on = 'subject') 
data_personal.email_response.nunique() # 44 unique completed responses
data_personal
```

```python
# now I need to remove ones with a weird name: subject ids 50,79,135,129, 170, 187, 197

not_name = [50, 79, 135, 129, 170, 187, 197]

data_personal=data_personal[-data_personal["subject"].isin(not_name)]


data_personal=data_personal.drop_duplicates()
data_personal=data_personal.drop_duplicates(subset=['subject', 'email_response', 'name_response'])
data_personal 
 # 45 pp left


data_personal = data_personal[data_personal.subject!= 133 ]
data_personal.subject.nunique()
data_personal.email_response.nunique()
```

```python
# extract email addresses
emails = data_personal.email_response.values.tolist()


emails = list(dict.fromkeys(emails))
#print(emails)


textfile = open("rewards2_s2.txt", "w")

for element in emails:
    textfile.write(element + "\n")
textfile.close() # these were sent the reward
```
